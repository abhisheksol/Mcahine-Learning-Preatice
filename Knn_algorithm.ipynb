```python
import pandas as pd
import seaborn as sns
from sklearn.datasets import load_iris
iris=load_iris()
iris
```

```python
iris.target_names   #target name mange flowers cha name ahe
iris.feature_names #columns cha names
# iris.data
```

```python
df=pd.DataFrame(iris.data,columns=iris.feature_names)
df
```

```python
df['target']=iris.target
df
```

```python
setosa=df[df.target==0]
versicolor=df[df.target==1]
virginica=df[df.target==2]

print(setosa)
print('----------------------------------------------------------')
print(versicolor)
print('------------------------\----------------------------------')
print(virginica)
print('----------------------------------------------------------')
```

```python
import matplotlib.pyplot as plt

# Assuming 'setosa', 'versicolor', and 'virginica' are pandas DataFrame objects with columns 'sepal_length' and 'sepal_width'

plt.figure(figsize=(10, 6))

# Plotting Setosa
plt.scatter(setosa['sepal length (cm)'], setosa['sepal width (cm)'], label='Setosa', color='red')

# Plotting Versicolor
plt.scatter(versicolor['sepal length (cm)'], versicolor['sepal width (cm)'], label='Versicolor', color='green')

# Plotting Virginica
plt.scatter(virginica['sepal length (cm)'], virginica['sepal width (cm)'], label='Virginica', color='blue')


plt.xlabel('Sepal Length')
plt.ylabel('Sepal Width')
plt.title('Sepal Length vs Sepal Width')
plt.legend()
plt.grid(True)
plt.show()
```
