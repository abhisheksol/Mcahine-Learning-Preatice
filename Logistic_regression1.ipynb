# In this logistic regression model, it predicts whether a given petal width belongs to the 'setosa' species or not.**(only by petal width) is trained**

```python
import seaborn as sns
import pandas as pd
import numpy as np
```

```python
df=sns.load_dataset('iris')
df.head()
# df.species.unique()
# df.isnull().sum()
```

```python
y = (df.species != 'setosa').astype(int) #convert false to 0 and vice-versa
x=df.petal_width
x
print(x,y)
```

```python
from sklearn.linear_model import LogisticRegression

# Reshape x to a 2D array with a single column
x_reshaped = x.values.reshape(-1, 1)

# Instantiate the Logistic Regression model
cf = LogisticRegression()

# Fit the model
cf.fit(x_reshaped, y)
```

# **Actual Prediction Here**

```python
example=cf.predict(([[0.8]]))
data=example[0]
if data == 0:
  print("Yes this is setosa ")
else:
  print("not a setosa")
```

# **Graph of Logistic Regression ,Where X is Width of Flower And Y is 0 || 1**

```python
import numpy as np
import matplotlib.pyplot as plt

# Generate x values for plotting
x_values = np.linspace(np.min(x), np.max(x), 100).reshape(-1, 1)

# Predict the probabilities for the generated x values
probabilities = cf.predict_proba(x_values)

# Plot the original data points
plt.scatter(x, y, color='blue', label='Original data')

# Plot the decision boundary
plt.plot(x_values, probabilities[:, 1], color='red', label='Logistic Regression')

# Add labels and legend
plt.xlabel('Petal Width')
plt.ylabel('Probability')
plt.title('Logistic Regression Decision Boundary')
plt.legend()

# Show the plot
plt.show()
```
